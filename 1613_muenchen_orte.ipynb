```python
import xml.etree.ElementTree as ET

# XML-Datei einlesen
file_path = '1613_muenchen.xml'
tree = ET.parse(file_path)
root = tree.getroot()

# TEI-Namespace definieren
namespaces = {'tei': 'http://www.tei-c.org/ns/1.0'}

# Alle Orte extrahieren
places = root.findall('.//*[@type="place"]', namespaces)

# Funktion zur Rekursion durch den XML-Baum
def extract_text_recursive(elem):
    return ''.join(elem.itertext()).strip()

# Orte extrahieren
all_places = {extract_text_recursive(elem) for elem in places if extract_text_recursive(elem)}

# Alphabetisch geordnete Wertausgabe
for place in sorted(all_places):
    print(place)
```
